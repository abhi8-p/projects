As the file is tab separted we use delimiter as '\t'

```python
df = pd.read_csv('/kaggle/input/life-expectancy-in-europe-by-age-and-sex/demo_mlexpec.tsv',delimiter = '\t')
```

```python
df.head()
```

```python
df.shape
```

```python
df.rename(columns={df.columns[0]:'new'},inplace = True)
```

```python
df.head()
```

SEPARATING DATA OF FIRST COLUMN

```python
df['unit'] = df['new'].apply(lambda x: x.split(',')[0])
df['sex'] = df['new'].apply(lambda x: x.split(',')[1])
df['age'] = df['new'].apply(lambda x: x.split(',')[2])
df['geo'] = df['new'].apply(lambda x: x.split(',')[3])
```

```python
df.drop('new',axis =1 ,inplace = True)
```

```python
df.head()
```

REORDERING THE COLUMNS

```python
col_list  = list(df.columns[-4:] )+list(df.columns[:-4])
df= df.reindex(columns = col_list)
```

```python
df.head()
```

Replacing **": "** To **NaN** values

```python
df = df.apply(lambda x : x.replace(df.iloc[0,4],np.NaN),axis = 1)
```

```python
df.head()
```

Removing text data from age, time columns

```python
df.age.unique()
```

```python
# filter out only numbers from the text data of age
age = []
def filter_only_numbers(x):
    age.append(''.join(filter(lambda x: x.isdigit(),x)))
df.age.apply(lambda x: filter_only_numbers(x))
df['age'] = age
```

```python
df.age.unique()
```

```python
#converts the time horizons to digits only, removes any character presennt
cols = df.columns[4:]
for x in cols:
    time = []
    def filter_only_numbers(x):
        time.append(''.join(filter(lambda x: x.isdigit() or x == '.',str(x))))
    df[x].apply(lambda x: filter_only_numbers(x))
    df[x] = time
```

```python
df.head()
```

```python
df.melt(id_vars=['unit','sex','age','geo'],var_name = "year",value_name = "value")
```

```python
data = df.copy(deep = False)
```

EXPORTING THE DATA

```python
data.drop('age',axis = 1,inplace = True)
data.to_csv('life_expectancy_eu.csv')
```
